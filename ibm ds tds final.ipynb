### Efficient calculation of Fibonacci sequence 

```python
import numpy as np

def fib_eff(k):
    f = round(((1+np.sqrt(5))**k - (1-np.sqrt(5))**k)/(np.sqrt(5)*2**k))
    return(f)

length = 100
for i in range(length):
    print(fib_eff(i))
```
